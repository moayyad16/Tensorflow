```python
# find min,max,min and sum of tensor
# make random tensor
S = tf.constant(np.random.randint(1 ,100 ,size=50),shape=(5,10))
S
```

```python
# find the min
tf.reduce_min(S)
```

```python
# find the max
tf.reduce_max(S)
```

```python
#find the mean
tf.reduce_mean(S)
```

```python
# find the sum
tf.reduce_sum(S)
```

```python
import tensorflow_probability as tfp
# find the veriancr and the variance is the measure of how far a set of numbers is spread out from their average value
tfp.stats.variance(S)
```

```python
# find the var throue the math lab
tf.math.reduce_variance(tf.cast(S, dtype=tf.float32))
```

```python
# find the standerd derviation
tf.math.reduce_std(tf.cast(S , dtype=tf.float32))
```
